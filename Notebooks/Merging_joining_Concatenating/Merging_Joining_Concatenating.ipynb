```python
import pandas as pd
```

```python
week1 = pd.read_csv(r"C:/Users/Willian/Desktop/Projects/Pandas_study/Data_set/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv(r"C:/Users/Willian/Desktop/Projects/Pandas_study/Data_set/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv(r"C:/Users/Willian/Desktop/Projects/Pandas_study/Data_set/Restaurant - Customers.csv")
foods = pd.read_csv(r"C:/Users/Willian/Desktop/Projects/Pandas_study/Data_set/Restaurant - Foods.csv")
```

# The pd.concat Method

```python
pd.concat(objs = [week1, week2], ignore_index = True)
```

```python
week1.append(other = week2, ignore_index = True)
```

```python
pd.concat(objs = [week1, week2], keys = ["Week 1", "Week 2"])
```

# Inner Joins

```python
week1.merge(week2, how = "inner", on = "Customer ID")
```

```python
week1.merge(week2, how = "inner", on = ["Customer ID", "Food ID"])
```

# Outer Joins

```python
week1.merge(week2, how = "outer", on = "Customer ID", indicator = True)
```

# Left Joins

```python
week1.merge(week2, how = "left", on = "Customer ID")
```
