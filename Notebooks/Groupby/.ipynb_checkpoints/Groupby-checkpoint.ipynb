```python
fortune = fortune.set_index("Rank")
```

```python
sectors = fortune.groupby("Sector")
```

# Retrieve a group with get_group Method

```python
sectors.get_group("Transportation")
```

```python
# We can combine get_group with group.size
```

```python
sectors.max()
sectors.min()
sectors.sum()
sectors.mean()
```

# Grouping by multiple columns

```python
groups2 = fortune.groupby(["Sector", "Industry"])
groups2.size()
```

# The .agg() Method

```python
sectors.agg({"Revenue":"sum",
            "Profits":"sum",
            "Employees":"mean"})
```

```python
sectors.agg(["size", "sum", "mean"])
```

# Iterating through groups

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue)
```

```python
df
```

```python
cities = fortune.groupby("Location")
```

```python
df2 = pd.DataFrame(columns = fortune.columns)
```

```python
df2
```

```python
for city, data in cities:
    highest_revenue = data.nlargest(1, "Revenue")
    df2 = df2.append(highest_revenue)
```
