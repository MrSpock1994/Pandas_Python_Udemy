```python
chicago = pd.read_csv(r"C:\Users\Willian\Desktop\Projects\Pandas_study\Data_set\chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head(3)
```

# Lower, upper, title and len

```python
chicago["Name"] = chicago["Name"].str.title()
```

```python
chicago["Position Title"] = chicago["Position Title"].str.title()
```

```python
chicago["Department"] = chicago["Department"].str.title()
```

```python
chicago
```

# The .str.replace() Method

```python
chicago["Department"] = chicago["Department"].str.replace("Mgmnt", "Management")
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$", "", regex = True).astype(float)
```

# Filtering with String Methods

```python
mask = chicago["Position Title"].str.contains("Water")
```

```python
chicago[mask]
```

```python
mask2 = chicago["Position Title"].str.startswith("Water")
```

```python
chicago[mask2]
```

```python
mask3 = chicago["Position Title"].str.endswith("ist")
```

```python
chicago[mask3]
```

# More String Methods - .strip(), .lstrip() and .rstrip()

```python
chicago["Name"] = chicago["Name"].str.strip()
```

```python
chicago["Position Title"] = chicago["Position Title"].str.strip()
```

# String Methods on Index and Columns labels

# Split Strings by Characters with .str.split() Method

```python
names.nunique()
names.value_counts()
```

```python
chicago[["First Name", "Last Name"]] = chicago["Name"].str.split(",", expand = True)
```
